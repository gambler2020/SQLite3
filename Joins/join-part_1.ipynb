```python
import sqlite3
```

```python
connection=sqlite3.connect("chinook.db")
```

```python
crsr=connection.cursor()
```

```python
#1. Using a subquery, find the names of all the tracks for the album "Californication".
```

```python
crsr.execute("select tracks.name from tracks where albumid in(select albums.albumid from albums where albums.title='Californication')")
```

```python
crsr.fetchall()
```

```python
#2. Find the total number of invoices for each customer along with the customer's full name, city and email.
```

```python
crsr.execute("select customers.firstname, customers.city, customers.email, (select count(*) from invoices where invoices.customerid=customers.customerid) as Number from customers")
```

```python
crsr.fetchall()
```

```python
#3. Retrieve the track name, album, artistID, and trackID for all the albums.
```

```python
crsr.execute("select tracks.name, albums.title, albums.artistid, tracks.trackid from albums left join tracks on albums.albumid=tracks.albumid")
```

```python
crsr.fetchall()
```

```python
#4. Retrieve a list with the managers last name, and the last name of the employees who report to him or her.
```

```python
crsr.execute("SELECT M.LastName AS Manager, E.LastName AS Employee FROM Employees E, Employees M WHERE E.ReportsTo = M.EmployeeID")
```

```python
crsr.fetchall()
```

```python
#5. Find the name and ID of the artists who do not have albums. 
```

```python
crsr.execute("SELECT Name, a.ArtistId, b.Title FROM Artists a LEFT JOIN Albums b ON a.ArtistId = b.ArtistId WHERE b.Title IS NULL")
```

```python
crsr.fetchall()
```

```python
#6. Use a UNION to create a list of all the employee's and customer's first names and last names ordered by the last name in descending order.
```

```python
crsr.execute("select employees.firstname, employees.lastname from employees union select customers.firstname, customers.lastname from customers order by lastname desc")
```

```python
crsr.fetchall()
```

```python
#7. See if there are any customers who have a different city listed in their billing city versus their customer city.
```

```python
crsr.execute("select customers.firstname, customers.lastname, customers.city from customers inner join invoices on customers.customerid=invoices.customerid where customers.city <> invoices.billingcity")
```

```python
crsr.fetchall()
```
