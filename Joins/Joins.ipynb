```python
import sqlite3
```

```python
connection=sqlite3.connect("chinook.db")
```

```python
crsr=connection.cursor()
```

```python
# 1. How many albums does the artist Led Zeppelin have? 
```

```python
crsr.execute("select artists.Artistid, Name as Artist from artists inner join albums on artists.Artistid=albums.Artistid where artists.Name='Led Zeppelin'")
```

```python
crsr.fetchall()
```

```python
crsr.execute("select count(Albumid) as TotalAlbums from albums where Artistid=(select Artistid from artists where Name='Led Zeppelin')")
```

```python
crsr.fetchall()
```

```python
# 2. Create a list of album titles and the unit prices for the artist "Audioslave".
```

```python
crsr.execute("select albums.Title, artists.Name, tracks.UnitPrice from((albums inner join artists on albums.artistid=artists.artistid) inner join tracks on tracks.albumid=albums.albumid) where artists.name='Audioslave'")
```

```python
crsr.fetchall()
```

```python
crsr.execute("select albums.Title, count(artists.Name), tracks.UnitPrice from((albums inner join artists on albums.artistid=artists.artistid) inner join tracks on tracks.albumid=albums.albumid) where artists.name='Audioslave'")
```

```python
crsr.fetchall()
```

```python
# 3. Find the first and last name of any customer who does not have an invoice. Are there any customers returned from the query? 
```

```python
crsr.execute("select customers.firstname, customers.lastname, count(invoices.invoiceid) from customers left join invoices on invoices.customerid=customers.customerid where invoices.invoiceid is null")
```

```python
crsr.fetchall()
```

```python
# 4. Find the total price for each album.
```

```python
crsr.execute("select albums.title, tracks.unitprice from albums inner join tracks on albums.albumid=tracks.albumid group by albums.title")
```

```python
crsr.fetchall()
```

```python
crsr.execute("select albums.title, sum(tracks.unitprice) from albums inner join tracks on albums.albumid=tracks.albumid group by albums.title")
```

```python
# 5. What is the total price for the album “Big Ones”?
```

```python
crsr.execute("select albums.title, sum(tracks.unitprice) from albums inner join tracks on albums.albumid=tracks.albumid where albums.title='Big Ones'")
```

```python
crsr.fetchall()
```
